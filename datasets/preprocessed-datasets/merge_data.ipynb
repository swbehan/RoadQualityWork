```python
alldata_fuel["Year"] = alldata_fuel["Year"].astype(int)
density_data["Year"] = density_data["Year"].astype(int)
all_data_quality["Year"] = all_data_quality["Year"].astype(int)

df_merged = pd.merge(alldata_fuel, density_data, on=["Country", "Year"], how="outer")
df_merged = pd.merge(df_merged, all_data_quality, on=["Country", "Year"], how="outer")
df_merged = df_merged.drop_duplicates()
df_merged = df_merged.drop(columns = ["Score_x", "Score_y", "Value"])
df_merged = pd.DataFrame(df_merged)
df_merged
```

```python
eu_iso_codes = {
    "Austria": "AUT",
    "Belgium": "BEL",
    "Bulgaria": "BGR",
    "Croatia": "HRV",
    "Cyprus": "CYP",
    "Czechia": "CZE",
    "Denmark": "DNK",
    "Estonia": "EST",
    "Finland": "FIN",
    "France": "FRA",
    "Germany": "DEU",
    "Greece": "GRC",
    "Hungary": "HUN",
    "Ireland": "IRL",
    "Italy": "ITA",
    "Latvia": "LVA",
    "Lithuania": "LTU",
    "Luxembourg": "LUX",
    "Malta": "MLT",
    "Netherlands": "NLD",
    "Poland": "POL",
    "Portugal": "PRT",
    "Romania": "ROU",
    "Slovakia": "SVK",
    "Slovenia": "SVN",
    "Spain": "ESP",
    "Sweden": "SWE"
}
df_merged['ISO'] = df_merged["Country"].map(eu_iso_codes)
```

```python
df_2019 = df_merged[df_merged['Year'] == 2019]
df_2021 = df_merged[df_merged['Year'] == 2021]
df_2024 = df_merged[df_merged['Year'] == 2024]
```

```python
def vis(df, hover_dct, text, year): 
    fig = px.choropleth(
        df,
        locations="ISO",
        locationmode="ISO-3",
        color="Country", 
        hover_data= hover_dct,
        scope='europe',
        color_discrete_sequence=px.colors.qualitative.Set3
    )

    fig.update_layout(
        margin=dict(l=10, r=10, t=35, b=0),
        width=800,
        height=500,
        title=f"EU countries in {year}: <br> {text}"
    )

    fig.show()
```
