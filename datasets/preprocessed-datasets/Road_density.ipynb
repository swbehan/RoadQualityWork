```python
def cleaned_data(data, year, name_list, EU_countries_list): 

    cleaned_dict = {}

    for item in data:
        if item.get('COMP_BREAKDOWN_1') == 'WEF_TTDI_VAL':
            ref_area = item.get('REF_AREA')
            obs_value = item.get('OBS_VALUE')
        if item.get('COMP_BREAKDOWN_1') == 'WEF_TTDI_SCR':
            obs_score = item.get('OBS_VALUE')
            if item.get('TIME_PERIOD') == year and ref_area not in cleaned_dict:
                cleaned_dict[ref_area] = year, obs_value, obs_score
    
    updated_dict = dict(zip(name_list, cleaned_dict.values()))

    filtered_dict = {k: v for k, v in updated_dict.items() if k in EU_countries_list}

    rows = [(k, v[0], v[1], v[2]) for k, v in filtered_dict.items()]
    df_country = pd.DataFrame(rows, columns=['Country', 'Year', 'Value', 'Score'])

    return df_country 
```

```python
data_24 = cleaned_data(data, '2024', names_list, EU_countries_list)
data_21 = cleaned_data(data, '2021', names_list, EU_countries_list)
data_19 = cleaned_data(data, '2019', names_list, EU_countries_list)

data_24
```

```python
def data_sorting(data): 
    data["Value"] = pd.to_numeric(data["Value"], errors="coerce")
    new_data = data.sort_values(by="Value")
    return new_data
```
