```python
def normalize_user_input(user_input):
    """
    Takes in user preferences on a 0-10 scale for:
    - fuel: importance on saving fuel prices
    - traffic: importance avoiding traffic
    - tourism: importance of avoiding high-tourism countries

    Returns: a normalized array on a 0-1 scale
    """
    max_scale = 10
    weights = np.array([
        user_input['fuel_price'] / max_scale,
        user_input['traffic_time'] / max_scale,
        user_input['tourism_num'] / max_scale
    ])
    return weights
```

```python
# example input
user_input = {
    'fuel_price': 8,
    'traffic_time': 5,
    'tourism_num': 9
}

user_weights = normalize_user_input(user_input)
print(user_weights)
```

```python
def compute_weighted_distace(row, weights):
    """
    Computes weighted distance between one country and the ideal distance vector ([0,0,0])
    """

    ideal_vector = np.array([0.0, 0.0, 0.0])
    
    country_vector = np.array([
        row["Fuel Score St."], 
        row["Density Score St."], 
        row["Tourism St."]
    ])
    diff = (country_vector - ideal_vector) * weights 

    return np.linalg.norm(diff)
```

```python
def get_top_5_recommendations(user_input, recommender_data): 
    """Reccomends top 5 countries closest to vector logo """

    user_weights = normalize_user_input(user_input)

    ideal_vector = np.array([0.0, 0.0, 0.0])

    recommender_data = recommender_data.copy()
    recommender_data["Distance"] = recommender_data.apply(
        lambda row: compute_weighted_distace(row, user_weights), axis = 1
    )

    top_5_recommendations = recommender_data.sort_values("Distance").head(5)
    return top_5_recommendations[["Country", "Distance"]]
```

```python
user_input = {
    "fuel_price": 0,
    "traffic_time": 5,
    "tourism_num": 10
}

top_5 = get_top_5_recommendations(user_input, recommender_data)
print(top_5)
```
